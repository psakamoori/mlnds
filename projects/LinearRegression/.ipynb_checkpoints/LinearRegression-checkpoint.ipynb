# Understanding Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('../../data/csv/heightVsWeight/hw_25000.csv')
```

```python
df.columns = ['Index', 'Height', 'Weight']
```

```python
df.head()
```

```python
df.plot(kind='scatter',
        x='Height',
       y='Weight',
       title='Weight and Height in adults')
```

```python
df.plot(kind='scatter',
        x='Height',
       y='Weight',
       title='Weight and Height in adults')
# Plotting a Line
plt.plot([61, 74], [85, 162], color='red', linewidth=4)
```

```python
def line(X, weight=0, bias=0):
    return X * weight + bias
```

```python
X = np.linspace(61, 74, 100)
```

```python
X
```

```python
yhat = line(X, weight=0, bias=0)
```

```python
yhat
```

```python
df.plot(kind='scatter',
        x='Height',
       y='Weight',
       title='Weight and Height in adults')
plt.plot(X, yhat, color='red', linewidth=4)
```

```python
def error(y_gt, y_pred):
    err = (y_gt - y_pred)
    return err.sum()
```

```python
def mse(y_gt, y_preed):
    s = (y_gt - y_pred)**2
    return s.mean()
```

```python
X = df[['Height']].values
y_gt = df[['Weight']].values
```

```python
y_pred = line(X)
```

```python
error(y_gt, y_pred)
```

```python
mse(y_gt, y_pred)
```
