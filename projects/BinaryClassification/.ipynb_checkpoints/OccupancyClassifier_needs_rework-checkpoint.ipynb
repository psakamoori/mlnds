# Classification

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam, SGD
```

```python
model = Sequential()
model.add(Dense(1, input_shape=(1,), activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(SGD(lr=0.5), 'binary_crossentropy', metrics=['accuracy'])
```

```python
ip = df['Light'].values
```

```python
label = df['Occupancy'].values
```

```python
model.fit(ip, label, epochs=500)
```

```python
gr = df.plot(kind='scatter', x='Light', y='Occupancy', title='Light Vs Occupancy')
temp_light = np.linspace(0, 1600)
gr.plot(temp_light, model.predict(temp_light), color='red')
plt.legend(['model', 'data'])
```

```python
test_df = pd.read_csv('../../data/txt/binary_classification_occupancy_data/datatest.txt')
```

```python
test_ip = test_df['Light'].values
gt_label = test_df['Occupancy'].values
gr = test_df.plot(kind='scatter', x='Light', y='Occupancy', title='Light Vs Occupancy')
gr.plot(test_ip, model.predict(test_ip), color='purple')
```
